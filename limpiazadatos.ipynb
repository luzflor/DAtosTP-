```python
#Agrego las columnas con valores unicos a "Columnas_a_borrar"
Columnas_a_borrar = ['Submitted_for_Approval','Last_Activity','ASP_(converted)_Currency','Actual_Delivery_Date','Prod_Category_A']
```

```python
#Verificamos si hay filas duplicadas
df.duplicated().sum()
```

```python
#Verificamos la canitdad de datos nulos de cada columnas
nullColumns = []    #Lista que contendr√° las columnas con valores Nulos
for column in df.columns:
    if df[column].isnull().sum() > 0:
        print(column, df[column].isnull().sum())
        nullColumns.append(column)
```

```python
df[nullColumns].isnull().sum().div(df.shape[0]).plot.bar();
```

```python
#Vemos la cantidad de repetici√≥n de los datos categ√≥ricos de ASP
df['ASP'].value_counts()
```

```python
#Usaremos el valor '0.38000' para reemplazar los 'nulls' de la columna
df.loc[df['ASP'].isnull(), 'ASP'] = 0.38
```

```python
#Verificamos que ya no hayan Nulos y seguimos
df['ASP'].isnull().sum() 
```

```python
#Vemos la cantidad de repetici√≥n de los datos categ√≥ricos de ASP
df['ASP_(converted)'].value_counts()
```

```python
#Usaremos el valor '0.38000' para reemplazar los 'nulls' de la columna
df.loc[df['ASP_(converted)'].isnull(), 'ASP_(converted)'] = 0.38
```

```python
#Verificamos que ya no hayan Nulos y seguimos
df['ASP'].isnull().sum() 
```

```python
#De la columna Total_Amount eliminamos los registros nulos
df = df[df['Total_Amount'].notnull()]

```

```python
#Chequeamos que no queden mas Nulos en todo el DataFrame:
df.isnull().sum().any()
```
