Analizando la estructura de la oportunidad se confirma que una oportunidad esta constituidad por varios item, estos se repiten por fecha estimado de entrega inicial y final.

Por otra parte se comprueba que el valor de monto gravado total es igual para todas las filas de la oportunidad comprobando que este es igual a la de los montos totales 

Buscar solo stage closed won y closed lost

```python
wl= df.loc[(df.Stage.str.contains('Closed Won')) | (df.Stage.str.contains('Closed Lost') ) ,:]
```

Filtrar solo con las columnas necesarias(region se utilizara en el proximo analisis)

```python
df_short=wl[['Region','Opportunity_Name','Stage']]
```

Elimanar duplicados y dejar solo una fila por oportunidad

```python
df_sin_duplicados = df_short.drop_duplicates() 
```

```python
Estados=df_sin_duplicados['Stage'].value_counts()
```

```python
# https://seaborn.pydata.org/generated/seaborn.barplot.html
g = sns.barplot(x=Estados.values, y=Estados.index, orient='h')
g.set_title("Cantidad de Oportunidades por Stage", fontsize=15)
g.set_xlabel("Cantidad de Oportunidades", fontsize=12)
g.set_ylabel("Stage", fontsize=12)
```

```python
g = sns.countplot(x='Region', hue='Stage', data=df_sin_duplicados, palette="hls", edgecolor='black', linewidth=0.5)
g.set_title('Oportunidades ganados y perdidas por Region', fontsize=20)
g.set_xlabel('Regiones', labelpad=15, fontsize=15)
g.set_xticklabels(g.get_xticklabels(), fontsize=11)
g.set_ylabel('Oportunidad', rotation=0, labelpad=50, fontsize=15)
```

```python
sum_per_each = pd.crosstab(df_sin_duplicados['Stage'],df_sin_duplicados['Region']).sum().to_frame().T
```

```python
sum_per_each
```

```python
df_sin_duplicados= pd.crosstab(df_sin_duplicados['Stage'], df_sin_duplicados['Region']).T
```

```python
for index, row in df_sin_duplicados.iterrows():
    df_sin_duplicados.loc[[index],:] =df_sin_duplicados.loc[[index],:].apply(lambda x: x / sum_per_each.at[0, index] * 100)

df_sin_duplicados
```

```python
ax = df_sin_duplicados.plot(kind='barh', stacked=True, figsize=(18,7), edgecolor='black', linewidth=0.5, width=0.4)

ax.set_title('Porcentaje de Ganados y Perdidos por Region', fontsize=20)

ax.set_ylabel('Regiones', rotation=0, labelpad=30, fontsize=15)
ax.set_xlabel('Porcentajes', rotation=0, labelpad=20, fontsize=15)

ax.set_yticklabels(ax.get_yticklabels(), fontsize=12)

ax.legend(loc='center left', bbox_to_anchor=(0.96, 0.5), ncol=1, fancybox=True, shadow=True, title='Porcentajes de Ganados y Perdidos por Region')

ax.title.set_position([0.5, 1.05])

plt.show()
```
